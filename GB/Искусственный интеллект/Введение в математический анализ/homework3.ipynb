### 1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

$\displaystyle \lim_{x \to \infty} f(x)$, где


|x| >= 1, x = $\sin(x)$ 

|x| < 1, x = $\sin(\frac{1}{x})$

Общий график функции

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    if np.fabs(x) < 1:
        return np.sin(1/x)
    else:
        return np.sin(x)
    
x = np.linspace(-10, 10, 200)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

### 2. Привести пример функции, не имеющей предела в точке, но определенной в ней.

$\displaystyle \lim_{x \to \infty} f(x)$, где

x > 0, x = $1$ 

x <= 0, x = $-1$

Общий график функции

```python
def f(x):
    if x > 0:
        return 1
    else:
        return -1
    
x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

$\displaystyle \lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}$

$\displaystyle \lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}$ = $\frac{3x^3 - 2x^2}{4x^2}$ = $\frac{x^2(3x - 2)}{4x^2}$ = $\frac{(3x - 2)}{4}$ = $-\frac{1}{2}$

#### b. 
$\displaystyle \lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}$

$\displaystyle \lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}$ = $\frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}$ = $\frac{x(\sqrt{1 + x} - \frac{1}{x})}{x(\sqrt[3]{1 + x} - \frac{1}{x})}$ = $\frac{\sqrt{1 + x} - \frac{1}{x}}{\sqrt[3]{1 + x} - \frac{1}{x}}$ = $\frac{\frac{1}{2}}{\frac{1}{3}}$ = $\frac{3}{2}$
