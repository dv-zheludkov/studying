### Условие 2: 2 задача


#### 2.1. Изучите распределение домов от наличия вида на набережную

Постройте график

Сделайте выводы


```python
plt.figure(figsize=(13, 13))
plt.pie(df['waterfront'].value_counts(), autopct='%1.1f%%')
plt.legend(['no', 'yes']);
plt.title('Распределение домов с видом на набережную');
```

#### 2.2. Изучите распределение этажей домов


```python
plt.figure(figsize=(13, 13))
plt.pie(df['floors'].value_counts(), autopct='%1.1f%%', labels=df['floors'].value_counts().index)
plt.title('Распределение домов по количеству этажей');
```

#### 2.3. Изучите распределение состояния домов

```python
plt.figure(figsize=(13, 13))
plt.pie(df['condition'].value_counts(), autopct='%1.1f%%', labels=df['condition'].value_counts().index);
plt.title('Распределение домов по состоянию');
```

```python
plt.figure(figsize=(13, 13))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Качество дома прямо влияет на стоимость.

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Распределение стоимости домов в зависимости от качества');
```

Наибольшую стоимость имеют дома с наилучшим качеством (13)

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Распределение стоимости домов в зависимости от количества этажей');
```

Большая часть наиболее дорогих домов имеет 2.5 этажа.

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Распределение стоимости домов в зависимости от вида');
```

Качество вида напрямую влияет на стоимость и наибольшую стоимость имеют дома с наилучшим видом.
